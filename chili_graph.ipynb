```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import re
import networkx as nx

%matplotlib inline
```

```python
df = pd.read_json("ingredient_co-occurance.json")
```

```python
G = nx.from_numpy_matrix(df.values)
G = nx.relabel_nodes(G, dict(enumerate(df.columns)))
```

```python
plt.figure(figsize=(40,40))
pos=nx.spring_layout(G, iterations=500, scale=5, k=.3)
nx.draw_networkx_labels(G,pos,fontsize=8)
font = {'fontname'   : 'Helvetica',
            'color'      : 'k',
            'fontweight' : 'bold',
            'fontsize'   : 14}
plt.title("Chili Graph", font)


nx.draw_networkx_nodes(G,pos,node_color='b',alpha=0.4)
nx.draw_networkx_edges(G,pos,alpha=0.4,node_size=0,width=1,edge_color='k')
plt.axis('off')
plt.show()
```
